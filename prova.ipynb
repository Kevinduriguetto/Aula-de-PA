```python
# Par ou ímpar Peça um número ao usuário e informe se ele é par ou ímpar.

# entrada

numero = int (input("digite um numero"))


# processamento / Saída

if numero % 2 == 0:

    print (f" o numero {numero} é par")

else:
    print (f"o numero {numero} e impar")
```

```python
# solicite dois números e mostre qual é o maior ( ou se são iguais)

# entrada
a = int(input("Digite o primeiro número inteiro: "))
b = int(input("Digite o segundo número inteiro: "))

# Processamento / Saída
if a > b:
    print(f"O número {a} é maior que {b}")
elif b > a:
    print(f"O número {b} é maior que {a}")
else:
    print(f"Os números {a} e {b} são iguais")
```

```python
# Peça a nota de um aluno e informe:
# Aprovado (nota ≥ 7)
 # Recuperação (5 ≤ nota < 7)
 # Reprovado (nota < 5)

 # Entrada

nota_bim = float(input("digite sua nota do bimestre "))


# Processamento / saida

if nota_bim < 5 :
  print ("Reprovado")
elif nota_bim < 7 :
  print("Recuperação")
else:
  print ("Aprovado")
```

```python
#Execicio 8 " Dia da semana Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo). "

print("--- Qual é o dia da semana? ---")
numero_dia = int(input("Digite um número de 1 a 7: "))
if numero_dia == 1:
    print("1 = Domingo")
elif numero_dia == 2:
    print("2 = Segunda-feira")
elif numero_dia == 3:
    print("3 = Terça-feira")
elif numero_dia == 4:
    print("4 = Quarta-feira")
elif numero_dia == 5:
    print("5 = Quinta-feira")
elif numero_dia == 6:
    print("6 = Sexta-feira")
elif numero_dia == 7:
    print("7 = Sábado")
else:
    print("Número inválido. Por favor, digite um número entre 1 e 7.")
```

```python
#Execicio 9 " Ano bissexto Solicite um ano e informe se ele é bissexto ou não. "
print("--- Verificador de Ano Bissexto ---")

# Solicita o ano ao usuário
ano = int(input("Digite um ano: "))

# Verifica se o ano é bissexto
# Um ano é bissexto se for divisível por 4, exceto se for divisível por 100
# Mas é bissexto se for divisível por 400
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é um ano bissexto.")
else:
    print(f"O ano {ano} NÃO é um ano bissexto.")
```
